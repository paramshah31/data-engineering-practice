```python
import requests,pandas as pd
```

```python
url = 'https://www.ncei.noaa.gov/data/local-climatological-data/access/2021/'
```

```python
response = requests.get(url)
response
```

```python
response.content
```

```python
data_in_list = pd.read_html(response.content)
```

```python
data_in_list
```

```python
df = data_in_list[0]
df.head()
```

```python
timestamp = '2022-02-07 14:03'
```

```python
df[df['Last modified'] == timestamp]['Name']
```

```python
len(df[df['Last modified'] == timestamp]['Name'])
```

### Let's get the 1st file

```python
filename = df[df['Last modified'] == timestamp]['Name'].iloc[0]
filename
```

```python
import wget
```

```python
wget.download(url+'/'+filename)
```

```python
actual_file = pd.read_csv(filename)
actual_file
```

```python
actual_file.columns
```

```python
max_drybulb_temp = actual_file['HourlyDryBulbTemperature'].max()
```
